```python
#import necessary libraries
import pandas as pd
import numpy as np
```

```python
#fetch the file
test2 = pd.read_csv ('energydata_complete (1).csv')
test = test2.set_index('date')
```

```python
test.head(2)
```

```python
#Firstly, we normalise our dataset to a common scale using the min max scaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
normalised_test =  pd.DataFrame(scaler.fit_transform(test), columns=test.columns)
features_test = normalised_test.drop(columns=[ 'lights' ])
target = normalised_test['Appliances']
```

```python
features_test.head(2)
```

```python
#Now, we split our dataset into the training and testing dataset. Recall that we had earlier segmented the features and target variables.

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_test, target, test_size= 0.3 , random_state= 42 )
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
#model = LinearRegression()
#fit the model to the training dataset

linear_model.fit(x_train, y_train)
#model.fit(x_train, y_train)
```

```python
#obtain predictions

predicted_values = linear_model.predict(x_test)
```

```python
#MAE

from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 3 )
```

```python
# _Residual Sum of Squares (RSS)_



rss = np.sum(np.square(y_test - predicted_values))
round(rss, 3 )
```

```python
# _Root Mean Square Error (RMSE)_

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3 )
```

```python
# _R-Squared_

from sklearn.metrics import r2_score

r2_score = r2_score(y_test, predicted_values)
round(r2_score, 3 )
```

```python
#_Ridge Regression_

from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha= 0.4 )
ridge_reg.fit(x_train, y_train)
```

```python

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3 )
```

```python
#_Feature Selection and Lasso Regression_

from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha= 0.001 )
lasso_reg.fit(x_train, y_train)
```

```python

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3 )
```

```python
#comparing the effects of regularisation

def get_weights_test(model, feat, col_name):
#this function returns the weight of every feature (the following codes should be in an indented block)
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_test = pd.DataFrame(weights).reset_index()
    weights_test.columns = [ 'Features' , col_name]
    weights_test[col_name].round(3)
    return weights_test
    return weights_test
```

```python
linear_model_weights = get_weights_test(linear_model, x_train, 'Linear_Model_Weight' )
ridge_weights_test = get_weights_test(ridge_reg, x_train, 'Ridge_Weight' )
lasso_weights_test = get_weights_test(lasso_reg, x_train, 'Lasso_weight' )
```

```python
lasso_weights_test
```

```python
final_weights = pd.merge(linear_model_weights, ridge_weights_test, on= 'Features' )
final_weights = pd.merge(final_weights, lasso_weights_test, on= 'Features' )

final_weights
```
